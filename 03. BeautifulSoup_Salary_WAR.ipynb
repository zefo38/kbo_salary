```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
year = [2020,2021,2022,2023,2024]
team = [2,1,6,9,5,3,7,12,13,11] # 팀 드롭다운 url code

# 총 Data를 담을 빈 list 생성
salary_data = []

for y in year :
    for t in team:
        url = f"https://statiz.sporki.com/add/?m=salary&year={y}&ct_code={t}"
        response = requests.get(url)
        html = response.text                               # text 속성 : UTF-8로 인코딩된 문자열을 받는다
        soup = BeautifulSoup(html, 'html.parser')          # BeautifulSoup 생성자로 html를 파싱 후 BS객체로 반환
        table = soup.find('div',{'class':'table_type03'})  
        headers = [header.text.strip() for header in table.find_all('th')] # 컬럼명으로 사용
        
        rows = [] # 선수 정보를 담을 빈 list 생성
        for row in table.find_all('tr')[1:]:
            cols = row.find_all('td')
            cols = [col.get_text(strip=True).replace('\n', '').replace('\r', '') for col in cols]
            rows.append(cols)
            
        salary_data.append(rows) # 한 팀의 연봉 데이터 추가
        
salary_data        
```

```python
# 3번 중첩된 리스트를 뽑아낸다
data = [data for salary in salary_data for data in salary ]
```

```python
Salary_WAR_data = pd.DataFrame(data, columns=headers)
```
