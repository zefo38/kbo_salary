## 4.2. 후년 연봉 컬럼 추가

```python
salary['후년연봉'] = salary['연봉(만원)']
```

```python
salary['연도'].unique()
```

```python
salary.info()
```

```python
salary
```

```python
sal = salary[['선수명','팀명','연도','후년연봉']]
sal['연도'] = (sal['연도'] - 1)
```

```python
sal.loc[sal['연도'] == 2020]
```

```python
all2020 = pd.merge(left=df2020, right=sal.loc[sal['연도'] == 2020], how='left',on=['선수명','팀명','연도' ],sort=False)
all2021 = pd.merge(left=df2021, right=sal.loc[sal['연도'] == 2021], how='left',on=['선수명','팀명','연도'],sort=False)
all2022 = pd.merge(left=df2022, right=sal.loc[sal['연도'] == 2022], how='left',on=['선수명','팀명','연도'],sort=False)
all2023 = pd.merge(left=df2023, right=sal.loc[sal['연도'] == 2023], how='left',on=['선수명','팀명','연도'],sort=False)
```

```python
all2020
```

```python
all2020.loc[all2020.duplicated(['선수명'], keep=False)].sort_values('선수명')
```

```python
# 2020년 stats과 연봉에 맞도록 들어가게 적절하지 않은 데이터 제거
# 이 경우 stats에는 없고 연봉 데이터에는 2명이 존재

all2020 = all2020.drop(index=126)
```

```python
# 중복으로 들어간 데이터를 선수 정보와 연봉에 맞게 들어가도록 특정 데이터 drop
all2022 = all2022.drop(index=[176,186,187])
```

```python
# 중복으로 들어간 데이터를 선수 정보와 연봉에 맞게 들어가도록 특정 데이터 drop
all2022 = all2022.drop(index=[215,216,217])
```
